# Step 4 : initializing the hyperparameters

```python
y.value_counts()
```

```python
# هنا رح نعطيه رقم التعلم لازم يكون اكبر مو الصفر و اقل من الواحد
learning_rate = 0.1
iterations = 5000
N = y_train.size

# Input features هنا عطيته رقم للمدخلات طبعا الكلوز الي عندي في الاكس
input_size = 7

# Hidden layers هنا عطيته رقم عشواء عشان يمشي عليها
hidden_size = 2

# Output layer هنا عطيته المخر من العمود التارقيت
output_size = 3
```

#Step 5 : Initialize Weights

```python
np.random.seed(10) # هنا الW
# على حسب ال hidden
# Hidden layer
W1=np.random.normal(scale=0.2,size=(hidden_size,output_size))
# Output layer
W2= np.random.normal(scale=0.5, size=(hidden_size , output_size))
```

#Step 6 : mean squared error, accuracy and sigmoid.



```python
# Helper functions
# كتابه الكود على شكل قانون
def sigmoid(x):# هنا احسبها بين الواحد و الصفر
    return 1 / (1 + np.exp(-x))

#هنا حسب نسبه الخطاء
def mean_squared_error(y_pred, y_true):# هنا اشوف نسبه الخطاء
    # One-hot encode y_true (i.e., convert [0, 1, 2] into [[1, 0, 0], [0, 1, 0], [0, 0, 1]])
    y_true_one_hot = np.eye(output_size)[y_true]

    # Reshape y_true_one_hot to match y_pred shape
    y_true_reshaped = y_true_one_hot.reshape(y_pred.shape)

    # Compute the mean squared error between y_pred and y_true_reshaped
    error = ((y_pred - y_true_reshaped)**2).sum() / (2*y_pred.size)#قانون

    return error
# هنا حسب الاكيورسي
def accuracy(y_pred, y_true):# هنا احسب الاكيورسي
    acc = y_pred.argmax(axis=1) ==  y_true.argmax(axis=1)
    return acc.mean()

# هنا حفظها في الداتا فيرم
results = pd.DataFrame(columns=["mse", "accuracy"])# احفظ القيم
```

#Step 7 : Building the Backpropogation Model in Python
###We will create a for loop for a given number of iterations and will update the weights in each iteration. The model will go through three phases feedforward propagation, the error calculation phase, and the backpropagation phase.


```python
# Training loop
# هنا بيمشي و يحسب ال باك برو

for itr in range(iterations):
    # Feedforward propagation
    Z1 = np.dot(X_train,W1)
    A1 = sigmoid(Z1)
    Z2 = np.dot(A1, W2)
    A2 = sigmoid(Z2)

    # Calculate error
    mse = mean_squared_error(A2, y_train)
    acc = accuracy(np.eye(output_size)[y_train], A2)
    new_row = pd.DataFrame({"mse": [mse], "accuracy": [acc]})
    results = pd.concat([results, new_row], ignore_index=True)

    # Backpropagation
    E1 = A2 - np.eye(output_size)[y_train]
    dW1 = E1 * A2 * (1 - A2)
    E2 = np.dot(dW1, W2.T)
    dW2 = E2 * A1 * (1 - A1)

    # Update weights
    W2_update = np.dot(A1.T, dW1) / N
    W1_update = np.dot(X_train.T, dW2) / N
    W2 = W2 - learning_rate * W2_update
    W1 = W1 - learning_rate * W1_update
```

```python
results.mse.plot(title="Mean Squared Error")
plt.show()
results.accuracy.plot(title="Accuracy")
plt.show()
#  رسم يوضح القيم
# نسبه الخطاء
# الاكيورسي له
```

###Now we will calculate the accuracy of the model.



```python
# Test the model
# اختبار المودل
Z1 = np.dot(X_test, W1)
A1 = sigmoid(Z1)
Z2 = np.dot(A1, W2)
A2 = sigmoid(Z2)
test_acc = accuracy(np.eye(output_size)[y_test], A2)
print("Test accuracy: {}".format(test_acc))
```
